```python
# import the ast library so that these list of dictionaries could be in a proper format
import ast
```

```python
# call the ast.literal_eval class and pass the output of one row. It will put it in a proper format.
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# create a function for doing the same task for every object and bring out only name from those dictionaries and put it in a list.
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
# call the function and pass the same output to check the function is working properly or not.
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# use apply function and pass the convert function in the 'genre' column. every row will be processed.
new_movies['genres'].apply(convert)
```

```python
# Assign the output in the same column
new_movies['genres'] = new_movies['genres'].apply(convert)
```

```python
# check the updated dataset
new_movies.head()
```

```python
# Apply the same function on the anathor similar column i.e., 'keywords'
new_movies['keywords'] = new_movies['keywords'].apply(convert)
```

```python
# create a function for cast column as we only want top 3 cast members from the movie
def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter = counter + 1
        else:
            break
    return l
```

```python
# apply the function in the cast column
new_movies['cast'] = new_movies['cast'].apply(convert3)
```

```python
# check the head of the data
new_movies.head()
```

```python
# Create a function in which it will process the data and only fetch director name from the crew members.
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
new_movies['crew'].apply(fetch_director)
```

```python
# Apply the function on crew column
new_movies['crew'] = new_movies['crew'].apply(fetch_director)
```

```python
'''We can see that, many words are similar and contains same meaning and 
they are repeating again and again. So we have to consider all those words as one word
eg: Love,Loves,Loving,Lovable etc. all these words will convert in a single word lov or luv
and it will use one word instead of all. For this we will use nltk library'''
```

```python
# We import the nltk library for NLP
import nltk
```

```python
# We install the necessery packages from nltk
from nltk.stem.porter import PorterStemmer
```

```python
# We call the PorterStemmer class
ps = PorterStemmer()
```

```python
# Create a function to pass the string for Natural Language Processing
def stemming(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
        
```

```python
# Checking the first row of tags column
new_df['tags'][0]
```

```python
# Putting the output of first row in stem function to check the difference in words
stemming('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
# apply the stem function on the tag column
new_df['tags'] = new_df['tags'].apply(stemming)
```

```python
# call the CountVectorizer and pass the values in parameters and assign it to a variable
cv = CountVectorizer(max_features=5000, stop_words= 'english')
```

```python
''' Fit and transform the tags column in CountVectorizer and convert the results to array and 
store it in a variable'''
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# Here we will see that similar words will not be repeating again
```

```python
'''get_feature_names will return the names of all the vectors and check the difference of 
before and now'''
cv.get_feature_names()
```

```python
'''Now we have the vectors, we will check the distance between vectors, the vectors 
which are near to each other will be similar to each other. For this we will use 
cosine similarity means the difference between angles of two vectors, if the angle will be more
the vectors will be far from each other, less angle difference means near to each other'''
```

```python
# import cosine similarity to check the difference of angles between the vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# pass the vectors in cosine similarity
cosine_similarity(vectors)
```

```python
# store it in a variable
similar_vectors = cosine_similarity(vectors)
```

```python
'''To get index number we will use enumerate function(it will return index no. and similarity in tuple)
it will sort on the basis of index no. but we want the sorting on the basis of similarity
So we will use a function which will take the input and will return its second value 
which is cosine similarity in our case and we will do indexing of second row to sixth row
for top 5 similar movies, starting from one because similarity of that perticular movie will always be 1
so we will start with second movie'''
sorted(list(enumerate(similar_vectors[0])),reverse=True,key = lambda x : x[1])[1:11]
```

```python
sorted(similar_vectors[0])[-10:-1]
```

```python
'''create a recommendation function to just try, is that working fine or not, 
it will return the movie index of top 5 related movies.'''
def recommend_try(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similar_vectors[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x : x[1])[1:11]
    
    for i in movies_list:
#         print(new_df.iloc[i[0]].title)
        print(i[0])
```

```python
# pass a movie in the function
recommend_try('Avatar')
```

```python
# check the movie name with movie index
new_df.iloc[1216].title
```

```python
# create a final function for recommendation which will return the name of the top 5 related movies
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similar_vectors[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x : x[1])[1:11]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
# We pass a movie to the function
recommend('Avatar')
```

```python
# We pass a movie to the function
recommend('Iron Man')
```
